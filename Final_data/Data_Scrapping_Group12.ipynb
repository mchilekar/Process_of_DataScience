```python
import os
import datetime
import pandas as pd
```

```python
date_records = pd.read_csv("/content/drive/MyDrive/Process_of_DataScience/Tweets/test_dates.csv")
date_records.head()
```

```python
dates_col = date_records["Date"]
required_dates = dates_col[:]
```

Here we have use snscrape library to extract the tweets from Twitter. We are extracting 1000 tweets for each date.


```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
keyword = "buy OR purchase OR sales OR electronics OR grocery OR furniture OR automatives "
query = keyword + "lang:en"
for d in required_dates:
  tweets = []
  file = 'tweet.csv'
  if(os.path.exists(file) and os.path.isfile(file)):
    os.remove(file)
  print(f"***Executing for date:{d}***")
  curr_date = datetime.datetime.strptime(d, '%Y-%m-%d').date()
  next_date = curr_date + datetime.timedelta(days=1)
  for i,tweet in enumerate(sntwitter.TwitterSearchScraper(keyword + 'since:'+d+' until:'+next_date.strftime('%Y-%m-%d')+' lang:en').get_items()):
    if i > 1000:
      break
    else:
      tweets.append([tweet.date, tweet.id, tweet.url, tweet.user.username, tweet.user.location, tweet.content, tweet.likeCount, tweet.retweetCount])
  df = pd.DataFrame(tweets,columns=['Date','Id','URL','username',"Location",'tweet','num_of_likes','num_of_retweet'])
  df["Date"] = df['Date'].dt.strftime('%Y.%m.%d')
  df.to_csv('tweet.csv',mode='a')
  df.to_csv("/content/drive/MyDrive/Process_of_DataScience/Tweets/tweet("+d+").csv",encoding='utf-8')

```
